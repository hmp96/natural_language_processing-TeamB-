```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_excel('data/감성대화말뭉치(원시데이터)_Training.xlsx')
final_data = pd.read_excel('data/감성대화말뭉치(최종데이터)_Training.xlsx')
```

```python
raw_data.head(3)
```

```python
final_data.head(3)
```

```python
raw_data = raw_data.iloc[:, 6:9]
final_data = final_data.iloc[:, 5:8]
```

```python
raw_data.head()
```

```python
final_data.head()
```

```python
raw_data.columns = ['sentiment1', 'sentiment2', 'dialog']
final_data.columns = ['sentiment1', 'sentiment2', 'dialog']
```

```python
raw_data.shape  # (74856, 3)
final_data.shape  # (40879, 3)
```

```python
len(set(raw_data['dialog']) & set(final_data['dialog']))  # 23837
set(raw_data['dialog']) & set(final_data['dialog'])
```

```python
train = pd.concat([raw_data, final_data]).reset_index(drop=True)
# train
```

```python
train = train.drop_duplicates().reset_index(drop=True)
```

```python
# train.csv로 파일 저장
train.to_csv('/content/drive/MyDrive/multicampus/team_project1/data/train.csv')
```
